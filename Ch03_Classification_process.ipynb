# 2. 이진 분류기 훈련
분류의 전체적인 프로세스에 관심을 가지고 보자. 
확률적 경사 하강법 분류기 SGDClassifier을 이용해보자.

```python
# 분류를 할 거니까 타겟 데이터를 범주형(True/False)로 바꾸자
y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5) # -> 넘파이 어레이 타입이다.(불린 벡터)
```

```python
from sklearn.linear_model import SGDClassifier
 
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)

sgd_clf.predict([some_digit]) # some_digit라는 1차원 어레이를 리스트에 담아서 준다.
# 결과는 5가 맞다고 했으므로 예측을 잘 했다!
```

# 3. 이진 분류기 성능 측정

## 3-1. 교차 검증을 사용한 정확도 측정

```python
# 교차 검증 과정을 직접 구현해서 제어해보자. 
# 다음 코드는 cross_val_score() 함수와 거의 같은 작업을 수행하고 동일한 결과를 출력합니다.
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

skfolds = StratifiedKFold(n_splits=3,shuffle=True ,random_state = 42)

for train_index, test_index in skfolds.split(X_train, y_train_5):
    print("이번에 뽑은 훈련/검증 셋 관측치의 인덱스 번호 : ",train_index, test_index)
    clone_clf = clone(sgd_clf)
    X_train_folds = X_train[train_index]
    print("이번에 뽑은 훈련 셋의 피쳐 : \n", X_train_folds)
    y_train_folds = y_train_5[train_index]
    print("이번에 뽑은 훈련 셋의 타겟(레이블) : ",y_train_folds)
    X_test_fold = X_train[test_index]
    print("이번에 뽑은 검증 셋의 피쳐 : \n",X_test_fold)
    y_test_fold = y_train_5[test_index]
    print("이번에 뽑은 검증 셋의 타겟(레이블)",y_test_fold)
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    print("검증 셋 피쳐를 보고 예측한 타겟(레이블) : ",y_pred)
    n_correct = sum(y_pred == y_test_fold) # 같은지 비교는 같은 인덱스 번호를 가진 관측치(동일한 관측치)끼리!
    print("예측값과 정답이 같은 관측치의 개수 : ",n_correct)
    print("정답을 맞춘 관측치 수/전체 관측치 수 : ", n_correct / len(y_pred))
    print('======================================')
```

```python
# cross_val_score() 함수를 이용해서 또 교차 검증해보자
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv =3, scoring='accuracy') # 분할 1번당 정확도 하나씩 나온다!
```

```python
# 무조건 '5아님'이라고 분류하는 더미 분류기를 만들어서 성능을 보자.(검증해보자)
from sklearn.base import BaseEstimator

class Never5Classifier(BaseEstimator) :
    def fit(self, X, y=None):
        return self
    def predict(self, X) : 
        return np.zeros((len(X),1), dtype = bool) # 영행렬을 만드는 메서드에 dtype=bool을 지정하면 0대산 False가 들어간 행렬을 준다.

never_5_clf = Never5Classifier()
cross_val_score(never_5_clf, X_train, y_train_5, cv =3, scoring='accuracy')
```

## 3-2. 오차 행렬

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
y_train_pred # 검증 셋 피쳐를 보고 예측한 타겟(레이블)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_5, y_train_pred)
#          예측        음                        양
# 실제 음[[5아닌 것을 5아닌 것으로 예측(TN), 5아님을 5로 예측(FP)],
#      양 [5인 것을 5 아니라고 예측(FN), 5인 것을 5로 예측(TP)]]
# (예측이 맞았는지 틀렸는지T/F + 예측값이 음인지 양인지N/P)
```

## 3-3. 정밀도와 재현율 

```python
from sklearn.metrics import precision_score, recall_score
print(precision_score(y_train_5, y_train_pred)) # 정밀도: 양성으로 예측한 것 중 진짜 양성이었던 것의 비율
print(recall_score(y_train_5, y_train_pred)) # 재현율/민감도 : 양성 샘플 중 양성이라고 감지한 비율
```

## 3-4. 정밀도와 재현율 트레이드오프

```python
y_scores = sgd_clf.decision_function(X_train) # decision_function()을 이용해 각 샘플의 결정함수의 점수를 얻을 수 있다.
y_scores
```

```python
X_train
```

```python
y_scores = sgd_clf.decision_function([some_digit]) # decision_function을 이용해 각 샘플의 결정함수의 점수를 얻을 수 있다.
for threshold in [0,4000]:
    y_some_digit_pred = (y_scores > threshold)
    print(y_some_digit_pred)
    
# 임계값을 높이면 진짜 5를 잡아내지 못하게 될 확률이 높아지므로 재현율이 떨어진다.
```

```python
# 모든 샘플의 결정함수 점수를 보자.  이때는 cross_val_predict를 써서 교차 검증 이용할 수 있다.
y_decision_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv =3, method = "decision_function")
y_decision_scores
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_decision_scores) # 정답 레이블과 결정함수 점수를 넣는다.
print("정밀도 : ",precisions,"\n","재현율 : ",recalls, "\n","결정임계값 : " ,thresholds) # 결정 임계값에 따른 정밀도와 재현율을 반환함
thresholds.shape # 근데 왜 결정 임계값 개수가 하나 더 적지??
```

```python
# PR(정밀도/재현율) 곡선 
def plot_precisions_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label = "precisions")
    plt.plot(thresholds, recalls[:-1], 'g-', label = "recalls")
    plt.legend()
    
plot_precisions_recall_vs_threshold(precisions, recalls, thresholds)
plt.show()
```

```python
# 재현율에 대한 정밀도 곡선 
plt.plot(recalls[:-1], precisions[:-1])
```

```python
# 정밀도가 90% 이상이되는 결정임계값을 찾는다.
import numpy as np
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]
```

```python
print(np.argmax(precisions >= 0.90)) # 정밀도가 90 이상에서 True가 처음 나오는 인덱스 번호를 반환 
precisions
```

```python
y_train_pred_90 = (y_decision_scores >= threshold_90_precision)
y_train_pred_90 # 앞서 구한 그 결정 임계값에 따라서 예측 레이블을 구함 
```

```python
# 정답과 예측 값을 넣어서 정밀도와 재현율 점수를 보자. 
print("모델의 정밀도 점수 : ",precision_score(y_train_5, y_train_pred_90)) # 정답 레이블과 예측 레이블을 넣음
print("모델의 재현율 점수 : ",recall_score(y_train_5, y_train_pred_90))
```

## 3-5. ROC(Receiver operating characteristic) 곡선 - TPR/FPR

```python
# TPR과 FPR 계산
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train_5, y_decision_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2,  label = label )
    plt.plot([0,1], [0,1], 'k--') # 대각 점선 표시
    
plot_roc_curve(fpr, tpr)
plt.show()
```

```python
# ROC 커브의 AUC(곡선 아래의 면적 ; Area Under the Curve)를 측정하면 분류기의 성능을 비교할 수 있습니다.
# 완벽한 분류기 : 1, 랜덤 분류기 : 0.5
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5, y_decision_scores)
```
